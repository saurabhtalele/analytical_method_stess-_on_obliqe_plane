```python
5#normal stress in x direction 
sigma_x= float(input("sigma_x to sign convenstions like -ve or +ve: "))

#normal stress in y direction 
sigma_y= float(input("sigma_y respect to sign convenstions like -ve or +ve : "))

#normal sher stress on plane xy 
#sign convention as per sign convention
tau_xy= float(input("tau_xy : "))

theta= float(input("theta n degrees with respect to sign convenstions like -ve or +ve : "))
theta=theta*(np.pi/180)


'''
for example below image and solution

sigma_x to sign convenstions like -ve or +ve: -80
sigma_y respect to sign convenstions like -ve or +ve : 50
tau_xy : -25
theta n degrees with respect to sign convenstions like -ve or +ve : -30
'''

'''
references   https://users.ox.ac.uk/~kneabz/Stress5_ht08.pdf
'''
```

half_solv.png

```python
# sigma_x_one stress on oblique cut plane
sigma_x_one=(((sigma_x+sigma_y)/2))+((((sigma_x-sigma_y)/2))*np.cos(2*theta))+((tau_xy)*np.sin(2*theta))
print("sigma_x_one : ",sigma_x_one)


# sigma_y_on stress on oblique cut plane
sigma_y_one=(((sigma_x+sigma_y)/2))-((((sigma_x-sigma_y)/2))*np.cos(2*theta))-((tau_xy)*np.sin(2*theta))
print("sigma_y_one : ",sigma_y_one)

# tau_x_one stress on oblique cut plane
tau_xy_one=-((((sigma_x-sigma_y)/2))*np.sin(2*theta))+((tau_xy)*np.cos(2*theta))
print("tau_xy_one : ",tau_xy_one)
```

on oblique plane.png

***finding principal plane angle and principal major and principal minor stress***

```python
# differentiate sigma_x_one with resp theta
theta_principale_dummy= (math.degrees(math.atan((tau_xy)/((sigma_x-sigma_y)/2))))

theta_principale_1=(theta_principale_dummy)/2

theta_principale_11=(np.pi/180)*theta_principale_1

theta_principale_2=(180+theta_principale_dummy)/2

theta_principale_22=(np.pi/180)*theta_principale_2

# sigma_x_one stress on oblique cut plane
sigma_one_principale=(((sigma_x+sigma_y)/2))+((((sigma_x-sigma_y)/2))*np.cos(2*theta_principale_11))+((tau_xy)*np.sin(2*theta_principale_11))

print("theta_principale_1 : ",theta_principale_1)
print("sigma_one : ",sigma_one_principale)


# sigma_y_on stress on oblique cut plane
sigma_two_principale=(((sigma_x+sigma_y)/2))+((((sigma_x-sigma_y)/2))*np.cos(2*theta_principale_22))+((tau_xy)*np.sin(2*theta_principale_22))
print("theta_principale_2 : ",theta_principale_2)
print("sigma_two_principale : ",sigma_two_principale)


```

```python
#finding principale share stess plane angle 
# differentiate tau_xy_one with resp theta
theta_principale_dummy_shear= (math.degrees(math.atan((-((sigma_x-sigma_y)/2))/(tau_xy))))
theta_principale_1_shear=(theta_principale_dummy_shear)/2

theta_principale_11_shear=(np.pi/180)*theta_principale_1_shear

theta_principale_2_shear=(180+theta_principale_dummy_shear)/2

theta_principale_22_shear=(np.pi/180)*theta_principale_2_shear
```

```python
tau_xy_one_1=-((((sigma_x-sigma_y)/2))*np.sin(2*theta_principale_11_shear))+((tau_xy)*np.cos(2*theta_principale_11_shear))
print("theta_principale_1_shear : ",theta_principale_1_shear)
print("tau_xy_one_1  : ",tau_xy_one_1)


tau_xy_one_2=-((((sigma_x-sigma_y)/2))*np.sin(2*theta_principale_22_shear))+((tau_xy)*np.cos(2*theta_principale_22_shear))
print("theta_principale_2_shear : ",theta_principale_2_shear)
print("tau_xy_one_2= : ",tau_xy_one_2)

normal_sigma_max_shear_plane=(sigma_x+sigma_y)/2
print("normal_sigma_max_shear_plane:",normal_sigma_max_shear_plane)
```
